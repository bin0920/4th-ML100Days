```python
i=0
for child in app_train['CNT_CHILDREN']:
    #print("i", i)
    if child == 0:
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="NO-CHILDREN"
        #print("child=0")
    elif child == 1:
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Few-CHILDREN"
    elif child == 2: 
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Few-CHILDREN"
        #print("child=2")
    elif child == 3: 
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Sev-CHILDREN"
    elif child == 4:
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Sev-CHILDREN"
    elif child == 5: 
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Sev-CHILDREN"
    else: 
        app_train.loc[i,['CNT_CHILDREN_GROUP']]="Man-CHILDREN"
    i+=1

list(app_train['CNT_CHILDREN_GROUP'])
```
